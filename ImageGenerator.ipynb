# Task
Generate an image from a text prompt using a pre-trained generative model.

**Reasoning**:
Import necessary classes and functions from the installed libraries for image generation.



```python
from diffusers import StableDiffusionPipeline
import torch
```

## Load the pre-trained model

### Subtask:
Load a pre-trained text-to-image model, such as Stable Diffusion, from a model hub like Hugging Face.


**Reasoning**:
Load the pre-trained Stable Diffusion model and move it to a CUDA device if available.



```python
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id)

if torch.cuda.is_available():
    pipe = pipe.to("cuda")

print("Model loaded successfully.")
```

## Generate image from text prompt

### Subtask:
Use the loaded model to generate an image based on a user-provided text prompt.


**Reasoning**:
Define a text prompt and use the loaded pipeline object to generate an image from the prompt.



```python
prompt = "a photo of horse in moon, cinematic, detailed, dramatic lighting"
image = pipe(prompt).images[0]
```

## Display the generated image

### Subtask:
Display the generated image.


**Reasoning**:
Display the image stored in the `image` variable.



```python
display(image)
```
